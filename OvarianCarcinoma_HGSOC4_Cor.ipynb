```python
cor1=cor[cor['cor']>0.8]
cor1=cor1[cor1['cor']<1]
cor1=pd.merge(cor1,index,left_on='cell1',right_on='index')
cor1=pd.merge(cor1,index,left_on='cell2',right_on='index')
cor1.columns=['cell1','cell2','cor','cell1index','cell1CB','cell2index','cell2CB']
cor2=cor1.iloc[:,[4,6,2]]
cor2
```

```python
data3=pd.read_table('/dssg/home/acct-medcl1/medcl1/Ovarian_carcinoma_GSE184880/HGSOC4/MT/OvarianCarcinoma_chrMmutation.csv',sep=',',index_col=0)
data3
```

```python
numchrM=data3['position'].value_counts()
numchrM=numchrM.to_frame().reset_index()
numchrM
```

```python
f=open('/dssg/home/acct-medcl1/medcl1/Ovarian_carcinoma_GSE184880/HGSOC4/MT/OvarianCarcinoma_tree08.csv','w')
for i in cor2.index:
    M=cor2.iloc[i,0]
    N=cor2.iloc[i,1]
    O=cor2.iloc[i,2]
    P=numchrM[numchrM['index'].isin([M])].iat[0,1]
    Q=numchrM[numchrM['index'].isin([N])].iat[0,1]
    if P - Q >0 :
        print ('%s\t%s\t%s\t%s\t%s/%s\t%s\t%s' %
                         (N, 
                          Q,
                          M,
                          P,
                          N,
                          M,
                          O,
                          1),file=f)
    else:
        print ('%s\t%s\t%s\t%s\t%s/%s\t%s\t%s' %
                         (M, 
                          P,
                          N,
                          Q,
                          M,
                          N,
                          O,
                          1),file=f)
f.close()
```
