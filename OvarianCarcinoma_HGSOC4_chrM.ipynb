```python
data=pd.read_table('/dssg/home/acct-medcl1/medcl1/Ovarian_carcinoma_GSE184880/HGSOC4/MT/OvarianCarcinoma_chrMall.csv',header=None,names=['position','number','ACTG','CB'])
data
```

```python
col=data.iloc[:,[0,1]]
col=col.drop_duplicates(keep='first')
A=pd.DataFrame(None,columns=['ACTG','number','position','total','frequency'])
for i in col['position']:
    mat=data.loc[data['position'] == i]
    D=mat.groupby(['ACTG']).size()
    D=D.to_frame().reset_index()
    D.columns=['ACTG','number']
    D['position']=i
    D['total']=col[col['position'].isin([i])].iat[0,1]
    A=A.append(D)
A['frequency']=A['number']/A['total']
```

```python
A.sort_values(by='position')
A
```

```python
f=open('/dssg/home/acct-medcl1/medcl1/Ovarian_carcinoma_GSE184880/HGSOC4/MT/OvarianCarcinoma_chrMmutationFRE.csv','w')
col=A['position']
col=col.drop_duplicates(keep='first')
for i in col:
    mat=A.loc[A['position'] == i]
    #D=mat['ACTG'].value_counts()
    #D=D.to_frame().reset_index()
    E=chrM[chrM['position'].isin([i])].iat[0,1]
    for j in range(len(mat['ACTG'])):
        if mat.iat[j,0] != E:
            print('%spos%s>%s\t%s' %
                 (i,
                 E,
                 mat.iat[j,0],
                 mat.iat[j,4]),file=f)
f.close()
```
