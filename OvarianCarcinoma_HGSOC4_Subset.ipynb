```python
import pandas as pd
```

```python
data=pd.read_table('/dssg/home/acct-medcl1/medcl1/Ovarian_carcinoma_GSE184880/HGSOC4/MT/OvarianCarcinoma_chrMdata.csv',sep=',',index_col=0)
data
```

```python
chrMfre=pd.read_table('/dssg/home/acct-medcl1/medcl1/Ovarian_carcinoma_GSE184880/HGSOC4/MT/OvarianCarcinoma_chrMmutationFRE.csv',header=None,names=['mutation','frequency'])
chrMfre
```

```python
chrMfre_partial=chrMfre[chrMfre['frequency']>0.1]
#chrMfre_partial=chrMfre_partial[chrMfre_partial['frequency']<0.9]
chrMfre_partial
```

```python
data1=pd.merge(data,chrMfre_partial,on='mutation')
data2=data1.iloc[:,[0,1,2]]
data2.columns=['mutation','frequency','position']
data2
```

```python
col=data2['position'].drop_duplicates(keep='first')
dict1={}
for i in col:
    mat=data2.loc[data2['position'] == i]
    mat1=mat['mutation'].values.tolist()
    mat1=list(set(mat1))
    dict1[i]=mat1
```

```python
f=open('/dssg/home/acct-medcl1/medcl1/Ovarian_carcinoma_GSE184880/HGSOC4/MT/OvarianCarcinoma_subset.csv','w')
def is_sub_set(short_list, long_list):
    set1, set2 = set(short_list), set(long_list)
    return set1.issubset(set2)
for i in dict1.keys():
    for j in dict1.keys():
        if i != j:
            if is_sub_set(dict1[i],dict1[j])==True:
                if dict1[i] == dict1[j]:
                    print ('%s/%s\t%s' %
                         (i, 
                          j,
                          0),file=f)
                if dict1[i] != dict1[j]:
                    print ('%s/%s\t%s' %
                         (i, 
                          j,
                          1),file=f)
f.close()
```
